```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# `Relational Plot`

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',size='size',hue='sex',style='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',size='size',hue='sex',style='time')
```

```python
gap = px.data.gapminder()
df = gap[gap['country'] == 'India']
df
```

```python
sns.lineplot(data=df,x='year',y='lifeExp') # axis level function

plt.show()
```

```python
sns.relplot(data=df,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
#facet plot

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',col='year',col_wrap=4)
```

# `Distribution Plot`

* histplot

```python
sns.histplot(data=tips,x='total_bill',kde=True)

plt.show()
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=15,hue='sex',element='step')

plt.show()
```
